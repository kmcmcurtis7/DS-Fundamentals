Because the counties really make this dataset messy and I am not using them in this analysis as I want to get a big pciture of what is going on in the datasets.

```python
# Removing counties and consolidating the data by grouping per state to the 2015 dataset.
state15 = df15.groupby(by='State').mean()

state15 = state15.reset_index(drop=False)
state15.head()
```

```python
# Removing counties and consolidating the data by grouping per state to the 2017 dataset.
state17 = df17.groupby(by='State').mean()

state17 = state17.reset_index(drop=False)
state17.head()
```

Since U.S. regions are referred to in many places from new articles, books, maps, etc, I decided that would be helpful in seeing the big picture easier.

```python
# Adding Regions column to 2015 dataset.
state15["Regions"] = state15["State"].replace(region)
state15.head()
```

```python
# Adding Regions column to 2017 dataset.
state17["Regions"] = state17["State"].replace(region)
state17.head()
```

```python
boxprops = dict(linestyle='-', linewidth=3)
medianprops = dict(linestyle='-', linewidth=3)

ax = state15.boxplot(figsize=(12,10),by='Regions', column='Women', grid=True, showmeans=True, 
                  showfliers=True,boxprops=boxprops,
                medianprops=medianprops)

# Removing automated generic title
plt.suptitle("")
ax.set_xlabel("")
ax.set_title("Total Population of Women by Region")
```

```python
w = state15.groupby('Regions')['Women'].sum().sort_values()
```

```python
w.plot(kind='bar', grid=True)

plt.ylabel('Average Population of Women')
plt.title('Average Population of Women per Region')
```

```python
boxprops = dict(linestyle='-', linewidth=3)
medianprops = dict(linestyle='-', linewidth=3)

ax = state17.boxplot(figsize=(12,10),by='Regions', column='Men', grid=True, showmeans=True, 
                  showfliers=True, boxprops=boxprops,
                medianprops=medianprops)

# Removing automated generic title
plt.suptitle("")
ax.set_xlabel("")
ax.set_title("Total Average Population of Men by Region")
```

```python
m = state15.groupby('Regions')['Men'].sum().sort_values()
```

```python
m.plot(kind='bar', grid=True)

plt.ylabel('Average Population of Men')
plt.title('Population Average of Men per Region')
```
