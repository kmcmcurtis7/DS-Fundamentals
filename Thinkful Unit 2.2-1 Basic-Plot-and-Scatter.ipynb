## Basic Plot and Scatter ##

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot([2, 4, 8, 16])
```

```python
# Make the random function consistent and replicable
np.random.seed(1221)

# Make a blank data frame
df = pd.DataFrame()

# Add a column of random numbers between 0 an 1
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2

# When creating a data frame an index column of counts is created, count starting at zero.
# Do a few transforms on that index to create some extra columns.
df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
# Now let's look at the random column
plt.plot(df['rand'])
plt.show()
```

` By default, plot creates a line graph with the index as the x-axis. `

### Basic Options ###
We are going to change the color to purple, expand the y-axis, label the y-axis, and give the whole thing a title.

```python
plt.plot(df['rand'], color='purple')
plt.ylim([-0.1, 1.1])
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

The logic for adjusting the x-axis is the same as the y, and there are countless other feature you can add to your charts.  Now let's add a second series.  To do this call `.plot( )` a second time for the second series.  Also adjust the y-axis.

```python
plt.plot(df['rand'], color='purple')
plt.plot(df['rand_shift'], color='green')
plt.ylim=([-0.1, 3.1])
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

### Scatter Plot ###
`.scatter()`

```python
plt.scatter(x=df['rand'], y=df['rand_sq'])
plt.show()
```

Let's use options to make the scatter plot more attractive.
<br>
Always feel free to look at the documentation, particularly the `matplotlib.pyplot.plot()` [documentation](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot) (see the section on keyword arguments) and other methods like [`.ylim()`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.ylim), [`.ylabel()`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.ylabel), and [`title()`](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.title) for more ideas. When in doubt, there's probably a way to do it.

```python
plt.scatter(
    x=df['rand'],
    y=df['rand_sq'],
    color='orange',
    marker='x', s=10
)
plt.scatter(
    x=df['rand'],
    y=df['rand_shift'],
    color='green',
    marker='x', s=10
    )
plt.show()
```

### Pandas Plot ###
While pandas can make some of this quicker, it isn't as powerful as matplotlib.<br>
Pandas give us the `.plot()` data frame method, so you can create a plot directly with pandas.  Though it is simpler to use, the plots aren't as pretty as pyplot functionality in matplotlib.

```python
df.plot(kind='scatter', x='counts_sq', y='counts_sqrt')
df.plot(kind='line')
plt.show()
```
