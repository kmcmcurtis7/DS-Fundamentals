## Statistical Plots ##

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Histogram ###
Great for showing the possible values of a variable and how common those values are.

Here is one form a nomral distribution.

```python
# Generate 1000 random values.
x = np.random.normal(10, 5, 1000)

# Plot them as a historgram
plt.hist(x)
plt.show()
```

The `10` and `5` we're passing into `random.normal()` are the **"center"** and the **"standard deviation"** of the normal curve we're using.  The histogram counts how many values are in each bin and plots a bar for that count for each bin. That gives you a sense of the density of the variable across its range. Histograms are fantastic for visualizing the distribution of a variable.

#### Bin count and placement####
By default, matplotlib distributes all values into ten(10) bins and chooses the best placement for those bins. That might not be the number you want and is adjustable with the `bins` keyword argument.

```python
# Random data
x = np.random.normal(10, 5, 1000)

# Create histogram and color
plt.hist(x, bins=40, color='purple')  # Changed bins from default of 10 to 40.
plt.title('Default Bin Placement')
plt.xlabel('Random Values')
plt.show()
```

```python
# Data to play with. Twice the histogram, twice the fun. :-)
x = np.random.normal(10, 5, 1000)
y = np.random.normal(15, 5, 10000)

# Override bind defaults with specific bin bounds
# FYI `alpha` controls the opacity.
plt.hist(x, color='blue', bins=np.arange(-10, 40), alpha=.5)
plt.hist(y, color='pink', bins=np.arange(-10, 40), alpha=.5)
plt.title('Manually setting bin placement')
plt.xlabel('Random Values')

plt.show()
```

### Normalizing Histograms ###

It's difficult to compare distributions this way.  **Normalization** allows you to solve this problem by _rescaling_ the height of the bins so that the total area under each curve sums to 1.

Useful when you have variables of different sizes (ie. hundreds vs thousands) and you wish to compare histograms.

```python
# Same data Normalized
plt.hist(x, normed=True, color='blue', bins=np.arange(-10, 40), alpha=.5)
plt.hist(y,normed=True, color='pink', bins=np.arange(-10, 40), alpha=.5)
plt.title('Normed histograms')
plt.xlabel('Random Values')

plt.show()
```

Much easier to compare as you can easily see the differences.

These are the basics of histograms in Python.  Explore the `.hist()`[documentation](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.hist) to learn more.

### Boxplot _(Also called box and whisker plot)_ ###  

Create boxplot with `plt.boxplot()` method.  This takes an array and a series of other options as parameters, as well as the standard pyplot options for things like color.

```python
# Set the random seed to keep the example consistent.
np.random.seed(111)

# Sample data
x = np.random.normal(10, 5, 1000)

# Generate and display the boxplot.
plt. boxplot(x)
plt.show()
```

This plot consists of a box, which covers what is called the interquartile range. Let's talk about quartiles for a moment. You can divide data into quarters like this:
 
 * the lowest value to the 25th percentile (Q1),
 * the 25th percentile to the median (Q2),
 * the median to the 75th percentile (Q3), and
 * the 75th percentile to the maximum value (Q4).

These are called _quartiles_, and the _middle two_ together are called the **_interquartile range_**, or **"_IQR_"**. So the box in our visualization includes all the data from the 25th percentile to the 75th percentile, with a red line indicating the median.

The whiskers are above and below the box. The whiskers will extend up and down up to a distance of 1.5x the IQR away from the top and bottom of the box. If all values are within that range, the whiskers will go to the max and min of the variable. Anything beyond 1.5x the IQR will be represented as a _flier_, or a small tick beyond the end of the whiskers. Our plot above has several fliers above and below the whiskers. Here is one with fliers on the top only.

```python
# Set the random seed to keep the example consistent.
np.random.seed(111)

# Sample Data
x = np.random.normal(10, 5, 100)

# Generate and display the box plot.
plt.boxplot(x)
plt.show()
```

Boxplots aren't useful for learning about the density or distribution of a variable. They are useful for investigating the range of values and identifying possible _outliers_ that warrant further investigation and may required special treatment.  For more information: `.boxplot()` [documentation](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.boxplot)
